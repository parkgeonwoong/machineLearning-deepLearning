```python
import numpy as np

data = [[2, 81], [4, 93], [6, 91], [8, 97]] # 일반적으로 데이터 묶여있음
x = [i[0] for i in data]
y = [i[1] for i in data]

# np.array 바꿔줘야함 -> 파이썬 배열은 벡터계산에 적합하지 않기 때문에
x_data = np.array(x)
y_data = np.array(y)

# 기울기와 절편 초기화
a = 0
b = 0
# 임의의 학습률
lr = 0.05
# 오차 줄이기위한 몇번 반복하는지
epochs = 2001

for i in range(epochs):
    y_pred = a * x_data + b     # 예측값
    error = y_data - y_pred     # 오차 = 실제 - 예측 
    y_err = ((y_data - y_pred)**2).mean()

    a_diff = -(1/len(x_data)) * sum(x_data * (error))   # MSE a로 미분
    b_diff = -(1/len(x_data)) * sum(error)              # MSE b로 미분
    # 경사하강법 공식
    a = a - lr * a_diff
    b = b - lr * b_diff

    if i % 100 == 0:
        print("epoch=%.f, 기울기=%.04f, 절편=%.04f, 오차=%.02f" % (i, a, b, y_err))
```
