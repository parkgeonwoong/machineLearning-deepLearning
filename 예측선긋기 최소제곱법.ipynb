```python
 import numpy as np

 x = [2, 4, 6, 8]
 y = [81, 93, 91, 97]
 
 # 평균 구하는 numpy => mean()
 mx = np.mean(x)
 my = np.mean(y)
 print('x의 평균: ', mx)
 print('y의 평균: ', my)
 
 # 공식 기울기 분모
 divisor = sum([(i-mx)**2 for i in x])

# 공식 기울기 분자
def top(x, mx, y, my):
    d = 0 
    for i in range(len(x)):
        d += (x[i] - mx) * (y[i] - my)
    return d

divided = top(x, mx, y, my)

print("분모: ", divisor)
print("분자: ", divided)

# 기울기=a, y절편= b 
a = divided / divisor
b = my - a*mx
print('a 기울기: ', a)
print('b 절편: ', b)
print("예측 그래프 : ", 'y = ',a,'x +', b)
```
