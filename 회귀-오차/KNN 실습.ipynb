## KNN 실습 - 학습 & 예측 

```python
import numpy as np
```

```python
d_data = np.column_stack((dachLength, dachHeight)) # 행으로 묶어서 데이터를 나타냄
d_label = np.zeros(len(d_data)) # 정답은 0이다. 
print("d_data: ",d_data)
print("d_label: ",d_label)
```

```python
s_data = np.column_stack((samoLength, samoHeight))
s_label = np.ones(len(s_data))
print(s_data)
print(s_label)
```

```python
newData = [[79, 35]]
```

```python
dog_classes = {0:'Dachshund', 1:'Samoyed'}
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
dogs = np.concatenate((d_data, s_data))
labels= np.concatenate((d_label, s_label))
print(dogs)
print(labels)
```

```python
k = 3 # 몇개의 이웃한 데이터를 가지고 할것인지 (주변에 3개 갖고 있으면 멈춤)
knn = KNeighborsClassifier(n_neighbors = k)
knn.fit(dogs, labels) # 여기 까지 학습
y_pred = knn.predict(newData) # 예측
print('예측값: ', y_pred)
print('데이터 ', newData, ', 판정 결과: ', dog_classes[y_pred[0]])
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier


# 데이터
dachLength = [77, 78, 85, 83, 73, 77, 73, 80]
dachHeight = [25, 28, 29, 30, 21, 22, 17, 35]

samoLength = [75, 77, 86, 86, 79, 83, 83, 88]
samoHeight = [56, 57, 50, 53, 60, 53, 49, 61]

# 닥스훈트 데이터 표현
d_data = np.column_stack((dachLength, dachHeight)) 
d_label = np.zeros(len(d_data))
#print(d_data)
#print(d_label)

# 사모예드 데이터 표현 
s_data = np.column_stack((samoLength, samoHeight))
s_label = np.ones(len(s_data))
#print(s_data)
#print(s_label)

# 새로운 데이터
newData = [[79, 30]]
# 클래스 구분
dog_classes = {0:"dach", 1:"samo"}

# 데이터 합쳐서 표현
dogs = np.concatenate((d_data, s_data))
labels = np.concatenate((d_label, s_label))
print(dogs)
print(labels)

# 학습 및 예측
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(dogs, labels)

y_pred = knn.predict(newData)
print(y_pred)
print('데이터 ', newData, ', 판정 결과: ', dog_classes[y_pred[0]])
```
