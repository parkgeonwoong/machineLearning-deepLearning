```python
import numpy as np

data = [[2, 0, 81], [4,4,93], [6,2,91], [8,3,97]]
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]

# np 배열로 바꾸기 => 인덱스로 하나하나 불러서 계산
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)

# 기울기, 절편, 학습률 초기화
a1 = 0
a2 = 0
b = 0
lr = 0.05
epochs = 2001

# 경사하강법
for i in range(epochs):
    y_pred = a1*x1_data + a2*x2_data + b
    error = y_data - y_pred
    y_err = ((y_data - y_pred)**2).mean()

    # 오차함수 미분
    a1_diff = -(1/len(x1_data)) * sum((error) * x1_data)
    a2_diff = -(1/len(x2_data)) * sum((error) * x2_data)
    b_diff = -(1/len(y_data) * sum(error))

    a1 = a1 - lr*a1_diff
    a2 = a2 - lr*a2_diff
    b = b - lr*b_diff

    if i % 100 == 0:
        print("epoch=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f, 오차=%.02f" % (i, a1, a2, b, y_err))
```
