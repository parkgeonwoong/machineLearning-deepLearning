```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10, 20, 30, 40])
plt.show()
```

```python
plt.plot([1,2,3,4], [12, 43, 25, 15])
plt.show()
```

```python
plt.title('plotting')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.title("Plotting")
plt.plot([10,20,30,40],color='skyblue', label="asc")
plt.plot([40,30,20,10], 'pink',  label="desc")
plt.legend() # asc, desc 보여주기
plt.show()
```

```python
plt.title("Marker")
plt.plot([10,20,30,40], 'r.', label="circle")
plt.plot([40,30,20,10], 'g^', label="Triangle")
plt.legend() 
plt.show()
```

```python
x = list(range(-10, 10))
print('x = ',x)

y = []
for i in x:
    y.append(2*i)
print('y = ',y)
print('')

plt.plot(x, y, marker='o')
plt.axis([-20, 20, -20, 20])
plt.title("y = 2x")
plt.show()
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67.0, 80.0 , 257.0, 1686.0, 6506.0, 11863.3, 22105.3]
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

plt.title("GDP per capita")
plt.ylabel("Dollars")
plt.xlabel("Years")
```
