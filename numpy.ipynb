```python
import numpy as np
```

```python
# 스칼라
x = np.array([3.0])
y = np.array([2.0])
print(x.shape) # 배열의 크기
print('x + y = ', x+y)
print('x * y = ', x*y)
```

```python
# 벡터
a = np.array([1,2, 3, 4, 5])
b = np.arange(4)
print(b)
```

```python
# 행렬
A = np.array([[1,2,3], [4, 5, 6]])
print(A)
A.shape # 두개 배열 3개 item

B = np.arange(20).reshape((5, 4))
print(B)
B.shape
```

```python
# 텐서
X = np.array([[[1, 2, 3], [4, 5, 6], [5, 6, 7]], [[1, 3, 4], [2, 3, 4], [6, 7, 8]]])
print(X)
X.shape

Y = np.arange(24).reshape((2, 3, 4))
print('Y.shape = ', Y.shape)
print('Y =', Y)
```

```python
# zeros(), eye(), ones() function
np.zeros((3, 4))
```

```python
np.ones((5, 2))
```

```python
# 브로드캐스팅
a = np.array([10, 11, 12])
print(a-10)
```

```python
A = np.array([[10, 20, 30], [14, 15, 16]])
print(A.shape)

print(A-a)
```

```python
# 벡터와 벡터의 곱셈 
x = np.array([1,2,3])
y =np.array([4, 5,6])

print(np.dot(x, y)) # 벡터 내적 dot
print(x@y) 
```

```python
# 행렬과 행렬의 곱셈
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[1,2],[3,4],[5,6]])
print(A.shape)
print(B.shape)
print(A@B)
```

```python
# 놈(Norm)
x = np.array([[1],[2],[3]])
L1_norm = np.linalg.norm(x, ord=1)
print(L1_norm)

L2_norm = np.linalg.norm(x, ord=2) # 제곱의 합 -> 제곱근
print(L2_norm)
```
