```python
import numpy as np

# 초기화 
w11 = np.array([-2, -2]) # 노드 n1으로 오는 값
w12 = np.array([2, 2]) # 노드 n2으로 오는 값
w2 = np.array([1,1]) # 은닉층에서 출력값으로 가는 값
b1 = 3
b2 = -1
b3 = -1

# 퍼셉트론 함수 
def MLP(x, w, b):
    y = np.sum(w*x) + b
    # 시그모이드 범위 0, 1 (활성화 함수)
    if y <= 0:
        return 0
    else:
        return 1

# 게이트 형태 표현
# NAND_gate 
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1) # np.array로 바꾸는 이유? -> 행렬 곱 사용하기 위해

# OR_gate
def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND_gate
def AND(x1, x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR_gate
def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))

# 데이터 값
for x in [(0,0), (1,0), (0,1), (1,1)]:
    y = XOR(x[0], x[1])
    print("입력 값: ", str(x), "출력 값: ", str(y))
```
