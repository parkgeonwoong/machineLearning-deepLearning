```python
import numpy as np 
import matplotlib.pyplot as plt

# 데이터 받기
data = [[2,0],[4,0],[6,0],[8,1],[10,1],[12,1],[14,1]]
x_data = [i[0] for i in data]
y_data = [i[1] for i in data]

# 기울기, 절편, 학습률 초기화 
a = 0
b = 0
lr = 0.01

# 시그모이드 함수 정의
def sigmoid(x):
    return 1 / (1 + np.e ** (-x))

# 경사하강법
for i in range(10001):
    for x_data, y_data in data: # 데이터의 값 하나하나씩 받아오기 위해서
        a_diff = x_data*(sigmoid(a*x_data + b) - y_data)    # 로그함수 오차공식 a 편미분
        b_diff = (sigmoid(a*x_data + b) - y_data)   # 로그함수 오차공식 b 편미분
        a = a - lr * a_diff     # 학습률
        b = b - lr * b_diff

        y_pred = sigmoid(a*x_data + b)  # 예측값 
        error = (y_data - y_pred) **2   # 오차의 제곱

        if i % 10 == 0:
            print("epoch=%.f, 기울기=%.04f, 절편=%.04f, y_data=%.04f ,예측값=%.04f, error=%.04f" % (i, a, b, y_data,y_pred, error))
```

```python
# 그래프 
plt.scatter(x_data, y_data)
plt.xlim(0, 15)
plt.ylim(-.1, 1.1)
```
