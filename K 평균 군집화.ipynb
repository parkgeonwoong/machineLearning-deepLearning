```python
import numpy as np
import matplotlib.pyplot as plt

dachLength = [77, 78, 85, 83, 73, 77, 73, 80]
dachHeight = [25, 28, 29, 30, 21, 22, 17, 35]

samoLength = [75, 77, 86, 86, 79, 83, 83, 88]
samoHeight = [56, 57, 50, 53, 60, 53, 49, 61]
```

```python
# 비지도 학습 --> 사모예드, 닥스훈트인지 모르고 해야한다.
dog_length = np.array(dachLength + samoLength)
dog_height = np.array(dachHeight+ samoHeight)

dog_data = np.column_stack((dog_length, dog_height))
# print(dog_data)

plt.title("Dog data without label")
plt.scatter(dog_length, dog_height)
plt.show()
```

```python
from sklearn import cluster
```

```python
model = cluster.KMeans(n_clusters=2)
model.fit(dog_data)
labels = model.predict(dog_data)
print(labels) # 서로 다른 범주 클러스터를 표현할 뿐 어디인지 모른다.
```

```python
plt.title("K-Means clustering")
colors = np.array(['red', 'green', 'blue', 'magenta'])
plt.scatter(dog_data[:, 0], dog_data[:, 1], color = colors[labels] )
plt.show()
```
